```python
import re
import pandas as pd

df= pd.read_csv(r'E:\UNMC Internship Document\Github Upload\data\dataset_live_links_v2.csv')
TLD2df = pd.DataFrame(columns=["url", "label"])

for i, j in zip(df["url"], df["label"]):
    doubleurl = re.findall(r"\.[a-zA-Z]{2}\b", i)

    if len(doubleurl) > 0:
        TLD2df.loc[len(TLD2df)] = [i, j]

TLD2df.to_csv('TLDLen2Filtered.csv')
```

```python
legitCounter=0
scamCounter=0
for i in TLD2df['label']:
    Label=re.findall(r"0",str(i))
    if len(Label)==1:
        legitCounter=legitCounter+1
    else:
        scamCounter=scamCounter+1

print(f"Legit Links: {legitCounter}")
print(f"Scam Links: {scamCounter}")
```
